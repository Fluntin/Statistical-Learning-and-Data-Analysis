```python
from scipy import stats
import pandas as pd
from scipy.special import gamma
from math import factorial, sin, cos, pi, comb
import numpy as np
import matplotlib.pyplot as plt
```

```python
α = 3 
β = 5


fig, ax = plt.subplots(1, 1)
x = np.linspace(stats.beta.ppf(0.0001, a=α, b=β),
                stats.beta.ppf(0.9999, a=α, b=β), 100)
rv = stats.beta(a=α, b=β)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='Beta(3,5)')
rvNorm = stats.norm(loc = α/(α+β), scale = ((α*β)/((α+β)**2*(α+β+1)))**(1/2))
ax.plot(x, rvNorm.pdf(x), 'k-', lw=2, color = 'red', label='Normal approx')

ax.legend(loc='best', frameon=False)
# plt.xlim([0,10])
# plt.ylim([0,1])

# save plot
# plt.savefig("gamma_pdf.pdf", bbox_inches='tight')

plt.show()
```

Normalfördelningen ger en bra approximering till betafördelningen när \alpha, \beta >= 10.

```python
xsum = sum([10, 10, 15, 16, 12, 11, 3, 4, 8, 5])
xsum
```

```python
#Aposteriori fördelningen är beta(γ,λ) där 
γ = 5 + xsum
λ = 3 + 20*10 - xsum
print(γ)
print(λ)
```

```python
# normalapproximationen är N(μ,σ^2) där
μ = γ/(γ+λ)
σ2 = (γ*λ)/((γ+λ)**2*(γ+λ+1))
print(μ)
print(σ2)
```
